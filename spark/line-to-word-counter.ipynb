```python
import csv
```

```python
#–í—Å–µ —á—Ç–æ str - —Ç–æ —Å–ª–æ–≤–æ :) 

with open('movies.csv', 'r') as mv:
    _dict = {}
    spamreader = csv.reader(mv, delimiter=',', quotechar='\n')
    for line in spamreader:
        
        init_line = line[1].lower().replace('"','')\
                                   .replace('.','')\
                                   .replace(':','')\
                                   .replace('(','')\
                                   .replace(')','')\
                                   .replace('-','')\
                                   .replace("'",'')\
                                   .replace("'s",'')\
                                   .split()
        for word in init_line:
            if word in _dict.keys():
                _dict[word] += 1
            else:
                _dict[word] = 1
                
    for w in sorted(_dict, key=_dict.get, reverse=True):
        print(w, _dict[w])
```
